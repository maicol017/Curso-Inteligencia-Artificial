```python
# prompt: creame 2 arreglos y matriz,el primero llamado saldo,que contiene datos sinteticos del saldo bancario,otro arreglo llamado antiguedad que que contiene datos de la antiguedad de la cuenta bancaria va de 0 a 10, una matriz de dos columnas que contiene la latitud en la columna 0, y la longitud en la columna 1, con datos muy cercanos a bucaramanga, es decir que varie la latitud y longitud de decimales o centecimales. el numero de datos son 20.la matriz se llama ubicacion.finalmente quiero que usando column stack me una las matrices y las listas creadas.defina la semilla aleatoria en 42

import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
saldo = np.random.randint(1000, 100000, size=20)
antiguedad = np.random.randint(0, 11, size=20)

# Generate latitude and longitude data near Bucaramanga (Colombia)
# Bucaramanga coordinates are approximately 7.125 N, 73.116 W
latitud = np.random.uniform(7.0, 7.3, size=20)
longitud = np.random.uniform(-73.3, -73.0, size=20)
ubicacion = np.column_stack((latitud, longitud))

# Stack the arrays and matrix
from numpy import column_stack

datos_combinados = column_stack((saldo, antiguedad, ubicacion))

print("Saldo:", saldo)
print("Antiguedad:", antiguedad)
print("Ubicacion:\n", ubicacion)
print("\nDatos combinados:\n", datos_combinados)
```

```python
print("Media",np.mean(datos_combinados[0]))
print("Max",np.max(datos_combinados[0]))
print("Min",np.min(datos_combinados[0]))
```

```python
# prompt: crea  un dataframe que una nombre,saldo,latitud,longitud y antiguedad

import pandas as pd
import numpy as np

# Generate synthetic data (assuming you want 20 rows as in the previous code)
nombres = np.array(['Sofia', 'Santiago', 'Valentina', 'Mateo', 'Camila', 'Samuel', 'Laura', 'Daniel', 'Isabella', 'Sebastian', 'Salome', 'Nicolas', 'Juan', 'Maria', 'Jose', 'Luisa', 'David', 'Gabriela', 'Alejandro', 'Valeria'])
saldo = np.random.randint(1000, 100000, size=20)
antiguedad = np.random.randint(0, 11, size=20)
latitud = np.random.uniform(7.0, 7.3, size=20)
longitud = np.random.uniform(-73.3, -73.0, size=20)

# Create the DataFrame
data = {'Nombre': nombres,
        'Saldo': saldo,
        'Latitud': latitud,
        'Longitud': longitud,
        'Antiguedad': antiguedad}
df = pd.DataFrame(data)

# Display the DataFrame
df
```
